# <p style="padding:10px;background-color:#7EC0EE;margin:0;color:white;font-family:newtimeroman;font-size:150%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Introduction üéì  </p>

- Most American institutions (that are not necessarily the top 10-20) that provide undergraduate education face a challenge i.e.
  - what kind of students they will make an offer for admission so that they can maintain a good performance in terms of number of students completing their courses in time.
  - They also need to predict how the current batch of intake will perform.
  
- The universities cannot relax the entry criteria too much as that way the quality of education that they provide gets diluted. At the other hand, they have to make offers to the candidates who are not only having appropriate profile but are also most likely to accept the offers.<br>

   ---

- On the other side, the students have a challenge deciding which colleges they should apply i.e.
   - the colleges that provide best performance at a minimal cost given their own profiles. <br>
The student profile is determined by not only the performance in examination such as SAT and ACT, but also other data points such as their ethnicity, immigration status, gender etc.

### 6. Do students prefer universities from highly `populated states`?

```python
New_data['State abbreviation'].unique()
```

```python
high_populated = ['California', 'Florida', 'Georgia', 'Illinois', 'Michigan','New York', 'North Carolina', 'Ohio', 'Pennsylvania', 'Texas']
```

```python
New_data['high_populated'] = New_data['State abbreviation'].apply(lambda x: "High" if x in high_populated else "Low")
```

```python
New_data.head().style.background_gradient(cmap='Blues')
```

```python
sns.barplot(data=New_data , x='high_populated' , y='Applicants total' ,  palette=[colors[0], colors[4]])
```

- it seems that students prefer High populated universities in application stage.

```python
sns.barplot(data=New_data , x='high_populated' , y='Admissions yield - total' ,  palette=[colors[0], colors[4]])
```

- but they tend to prefer Low populated in Enrollment stage.
---

### 7. Do students prefer a university because it belongs to a `state with a high GDP per capita`?

```python
high_GDP = ['Massachusetts', 'Connecticut', 'New York', 'California', 'New Jersey','Alaska','Texas']
```

```python
New_data['high_GDP'] = New_data['State abbreviation'].apply(lambda x: "High" if x in high_GDP else "Low")
```

```python
New_data.head().style.background_gradient(cmap='Blues')
```

```python
sns.barplot(data=New_data , x='high_GDP' , y='Applicants total' ,  palette=[colors[0], colors[4]])
```

- it seems that students prefer High GDP universities in application stage.

```python
sns.barplot(data=New_data , x='high_GDP' , y='Admissions yield - total' ,  palette=[colors[0], colors[4]])
```

- but students prefer Low GDP universities in enrollment stage.
---

### 8. Do students prefer a university based on the possibility of a `higher, additional academic degree` in the same university?

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=New_data , x='Highest degree offered' , y='Applicants total' ,  palette=[colors[0],colors[1],colors[2],colors[3],colors[4]])
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=New_data , x='Highest degree offered' , y='Admissions yield - total' , palette=[colors[0],colors[1],colors[2],colors[3],colors[4]])
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
```

- it seems that it doesnot matter in enrollment, but they prefer higher degree while application stage.
