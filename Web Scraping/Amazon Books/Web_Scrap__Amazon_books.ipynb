```python
from selenium import webdriver

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.headless = True

katkoot = webdriver.Chrome(options=chrome_options)
```

```python
katkoot.get("https://www.amazon.in/gp/bestsellers/books/")
```

```python
web_elements =katkoot.find_elements("xpath","//*[@id='zg-right-col']")
```

```python
for i in web_elements :
  print(i.text)
```

```python
import math

title_list = []
price_list = []

for book in web_elements:

    title_elements = book.find_elements("xpath", ".//a/span/div")
    if title_elements:
      for title_element in title_elements:
          title_list.append(title_element.text)
    else:
      title_list.append(math.nan)


    price_elements = book.find_elements("xpath", ".//span[contains(@class, 'p13n-sc-price')]")
    if price_elements:
      for price_element in price_elements:
          price = price_element.text.replace('₹', '').strip()
          price_list.append(price)
    else:
      price_list.append(math.nan)
```

```python
import pandas as pd
data = pd.DataFrame({
    'Title': title_list,
    'Price': price_list
})
data
```

```python
data.to_csv("book_data.csv",index=False)
```
