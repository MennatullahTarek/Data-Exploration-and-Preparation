```python
from selenium import webdriver

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.headless = True

katkoot = webdriver.Chrome(options=chrome_options)
```

```python
katkoot.get("https://www.amazon.eg/s?k=samsung&rh=p_89%3Asamsung&language=en&ref=SQEG-WEB-SR301")
```

```python
web_elements=katkoot.find_elements("xpath","//div[@class='a-section a-spacing-base']")
```

# <p style="padding:10px;background-color:#7EC0EE;margin:0;color:white;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 50px 50px;overflow:hidden;font-weight:500"> 2. Extract data :</p>

```python
import re
import pandas as pd
import math
```

```python
product_list = []

product_pattern =  r"(SAMSUNG)"

for element_text in web_elements:
  match = re.search(product_pattern, element_text.text)
  if match:
    product = match.group(1)
    product_list += [product]

  else:
    product_list.append(math.nan)
```

```python
model_list = []

model_pattern = r"(Galaxy\s[A-Za-z0-9]+)"

for element_text in web_elements:
  match = re.search(model_pattern, element_text.text)
  if match:
    model = match.group(1)
    model_list += [model]

  else:
    model_list.append(math.nan)
```

```python
OS_list = []

OS_pattern =  r"(Android|iOS|Windows|Linux|MacOS)"

for element_text in web_elements:
  match = re.search(OS_pattern, element_text.text)
  if match:
    OS = match.group(1)
    OS_list += [OS]

  else:
    OS_list.append(math.nan)
```

```python
storage_list = []

storage_pattern = r"\d+GB Storage"

for element_text in web_elements:
  match = re.search(storage_pattern, element_text.text)
  if match:
    storage = match.group(0)
    storage_list += [storage]

  else:
    storage_list.append(math.nan)
```

```python
ram_list = []

ram_pattern = r"\d+GB RAM"

for element_text in web_elements:
  match = re.search(ram_pattern, element_text.text)
  if match:
    ram = match.group(0)
    ram_list += [ram]

  else:
    ram_list.append(math.nan)
```

```python
color_list = []

color_pattern = r"\b(?:Black|Silver|Blue|Green|Navy|Iceblue|Lemon|Lilac|Gray|Graphite|Titan|Gold|Yellow|Purple|Clean White|Inox|Multi Color)\b"


for element_text in web_elements:
  match = re.search(color_pattern, element_text.text)
  if match:
    color = match.group(0)
    color_list += [color]

  else:
    color_list.append(math.nan)
```

```python
price_list = []

price_pattern = r"EGP\s*([\d,.]+)"

for element_text in web_elements:
  match = re.search(price_pattern, element_text.text)
  if match:
    price_str = match.group(1)
    price_str = price_str.replace(",", "")
    try:
        price = float(price_str)
        price_list += [price]
    except ValueError:
        print("Invalid price format:", price_str)
  else:
    price_list.append(math.nan)
```

```python
purchases_list = []

purchases_pattern = r"(\d+\+?\s?(bought|offers|sold))"

for element_text in web_elements:
  match = re.search(purchases_pattern, element_text.text)
  if match:
    purchases = match.group(1)
    purchases_list += [purchases]

  else:
    purchases_list.append(math.nan)
```

```python
delivery_date_list = []

delivery_date_pattern = r"(?:Get it by|Get it as soon as|FREE delivery|delivery)\s*(?:\w+,\s*)?\s*(\d{1,2}\s\w+)"

for element_text in web_elements:
    matches = re.findall(delivery_date_pattern, element_text.text)
    if matches:
        delivery_date_list += matches
    else:
       delivery_date_list.append(math.nan)
```

```python
data = pd.DataFrame({
    'Product': product_list,
    'Model': model_list,
    'OS': OS_list,
    'Storage': storage_list,
    'RAM': ram_list,
    'Color': color_list,
    'Price (EGP)': price_list,
    'Purchases': purchases_list,
    'Delivery Date': delivery_date_list
})
data
```
